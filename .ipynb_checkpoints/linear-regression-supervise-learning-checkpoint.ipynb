```python
import pandas as pd

# AAP daily OHLCV; path is relative to this notebook's working directory
csv_path = "./individual_stocks_5yr/AAP_data.csv"
df = pd.read_csv(csv_path)

# Quick sanity check: column names, dtypes, first rows
df.head()
```

```python
# Null/NaN vs empty strings are separate: "" is not NaN in object columns
null_per_col = df.isna().sum()
print("Null / NaN count per column:")
print(null_per_col.to_string())

cols_with_nulls = null_per_col[null_per_col > 0]
print()
if len(cols_with_nulls):
    print("Columns with at least one missing value:")
    print(cols_with_nulls.to_string())
else:
    print("No null or NaN in any column (all counts are 0).")

# Empty = whitespace-stripped string length 0 (skip numeric/datetime dtypes)
def _n_empty_strings(col):
    if pd.api.types.is_numeric_dtype(col) or pd.api.types.is_datetime64_any_dtype(col):
        return 0
    return col.astype(str).str.strip().eq("").sum()

empty_per_col = df.apply(_n_empty_strings)
print()
print("Empty string count per column (after strip):")
print(empty_per_col.to_string())

cols_with_empty = empty_per_col[empty_per_col > 0]
print()
if len(cols_with_empty):
    print("Columns with at least one empty string:")
    print(cols_with_empty.to_string())
else:
    print("No empty strings in scanned text columns (all counts are 0).")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Univariate linear regression: ONE input feature (today's close) -> next day's close.
# Two parameters: intercept + slope on standardized today's close (contour plot is 2D in those).
df_supervised = df.copy()
df_supervised["close_next"] = df_supervised["close"].shift(-1)
df_supervised = df_supervised.dropna(subset=["close_next"])

target_next_day_close = df_supervised["close_next"].to_numpy().reshape(-1, 1)
feature_today_close = df_supervised["close"].to_numpy().reshape(-1, 1)
n_samples = target_next_day_close.shape[0]

# Standardize the single feature so learning_rate stays in a sensible range
today_close_mean = feature_today_close.mean()
today_close_std = feature_today_close.std()
today_close_standardized = (feature_today_close - today_close_mean) / today_close_std
design_matrix = np.hstack(
    [np.ones((n_samples, 1)), today_close_standardized]
)

learning_rate = 0.01  # adjust manually
num_iterations = 2000

model_parameters = np.zeros((2, 1))
gradient_descent_parameter_history = np.zeros((num_iterations + 1, 2))
gradient_descent_parameter_history[0] = model_parameters.ravel()
cost_history = []

for iteration in range(num_iterations):
    predictions = design_matrix @ model_parameters
    residuals = predictions - target_next_day_close
    cost_history.append(float((residuals.T @ residuals) / (2 * n_samples)))
    gradient = (design_matrix.T @ residuals) / n_samples
    model_parameters = model_parameters - learning_rate * gradient
    gradient_descent_parameter_history[iteration + 1] = model_parameters.ravel()

final_residuals = design_matrix @ model_parameters - target_next_day_close
cost_history.append(float((final_residuals.T @ final_residuals) / (2 * n_samples)))

closed_form_parameters = np.linalg.inv(design_matrix.T @ design_matrix) @ design_matrix.T @ target_next_day_close


def mean_squared_error_cost(intercept, slope_on_standardized_feature):
    parameter_vector = np.array([[intercept], [slope_on_standardized_feature]])
    errors = design_matrix @ parameter_vector - target_next_day_close
    return float((errors.T @ errors) / (2 * n_samples))


closed_intercept = float(closed_form_parameters[0, 0])
closed_slope = float(closed_form_parameters[1, 0])
intercept_span = max(8.0, abs(closed_intercept) * 0.4 + 1e-6)
slope_span = max(0.8, abs(closed_slope) * 0.4 + 1e-6)
intercept_axis = np.linspace(closed_intercept - intercept_span, closed_intercept + intercept_span, 90)
slope_axis = np.linspace(closed_slope - slope_span, closed_slope + slope_span, 90)
intercept_mesh, slope_mesh = np.meshgrid(intercept_axis, slope_axis)
cost_on_grid = np.array(
    [
        [
            mean_squared_error_cost(intercept_mesh[i, j], slope_mesh[i, j])
            for j in range(intercept_mesh.shape[1])
        ]
        for i in range(intercept_mesh.shape[0])
    ]
)

fig_contour, ax_contour = plt.subplots(figsize=(7, 5.5))
contour_set = ax_contour.contour(intercept_mesh, slope_mesh, cost_on_grid, levels=25)
fig_contour.colorbar(contour_set, ax=ax_contour, label=r"$J(\theta)$")
ax_contour.plot(
    gradient_descent_parameter_history[:, 0],
    gradient_descent_parameter_history[:, 1],
    "r.-",
    markersize=3,
    label="gradient descent",
)
ax_contour.plot(closed_intercept, closed_slope, "g*", markersize=14, label="closed form")
ax_contour.set_xlabel(r"$\theta_0$ (intercept)")
ax_contour.set_ylabel(r"$\theta_1$ (slope on standardized today's close)")
ax_contour.set_title("Cost contours + GD path (univariate: one feature)")
ax_contour.legend()
plt.tight_layout()
plt.show()

fig_cost, ax_cost = plt.subplots(figsize=(7, 3.5))
ax_cost.plot(cost_history, color="steelblue")
ax_cost.set_xlabel("iteration")
ax_cost.set_ylabel(r"$J(\theta)$")
ax_cost.set_title("Cost vs iteration")
plt.tight_layout()
plt.show()

today_close_line_axis = np.linspace(float(feature_today_close.min()), float(feature_today_close.max()), 200)
standardized_line_axis = (today_close_line_axis - today_close_mean) / today_close_std
predicted_next_day_close_line = (
    model_parameters[0, 0] + model_parameters[1, 0] * standardized_line_axis
)
fig_fit, ax_fit = plt.subplots(figsize=(7, 5))
ax_fit.scatter(feature_today_close.ravel(), target_next_day_close.ravel(), s=10, alpha=0.35, label="data")
ax_fit.plot(
    today_close_line_axis,
    predicted_next_day_close_line,
    color="darkorange",
    lw=2,
    label="linear regression (GD)",
)
ax_fit.set_xlabel("Today's close (day t)")
ax_fit.set_ylabel("Next day's close (day t+1)")
ax_fit.set_title("Fitted regression line (one feature)")
ax_fit.legend()
plt.tight_layout()
plt.show()

print(f"learning_rate = {learning_rate}, num_iterations = {num_iterations}")
print(f"parameters (GD):     {model_parameters.ravel()}")
print(f"parameters (closed): {closed_form_parameters.ravel()}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# --- Multivariate: open, high, low, close, volume -> close_next (same-day row, time-ordered split) ---
feature_column_names = ["open", "high", "low", "close", "volume"]
df_multi = df.copy()
df_multi["close_next"] = df_multi["close"].shift(-1)
df_multi = df_multi.dropna(subset=["close_next"])

features_raw = df_multi[feature_column_names].to_numpy()
target_next = df_multi["close_next"].to_numpy().reshape(-1, 1)

features_train_raw, features_test_raw, target_train, target_test = train_test_split(
    features_raw,
    target_next,
    test_size=0.2,
    shuffle=False,
)

n_train_samples = features_train_raw.shape[0]
feature_means = features_train_raw.mean(axis=0)
feature_stds = features_train_raw.std(axis=0)
features_train = (features_train_raw - feature_means) / feature_stds
features_test = (features_test_raw - feature_means) / feature_stds

design_train_multi = np.hstack(
    [np.ones((n_train_samples, 1)), features_train]
)
design_test_multi = np.hstack([np.ones((features_test.shape[0], 1)), features_test])
n_parameters = design_train_multi.shape[1]

learning_rate_multi = 0.05
num_iterations_multi = 4000
parameters_multi = np.zeros((n_parameters, 1))
training_cost_history = []

for _ in range(num_iterations_multi):
    predictions_train = design_train_multi @ parameters_multi
    residuals_train = predictions_train - target_train
    training_cost_history.append(
        float((residuals_train.T @ residuals_train) / (2 * n_train_samples))
    )
    gradient_multi = (design_train_multi.T @ residuals_train) / n_train_samples
    parameters_multi = parameters_multi - learning_rate_multi * gradient_multi

predictions_train_final = design_train_multi @ parameters_multi
predictions_test = design_test_multi @ parameters_multi

r2_train = r2_score(target_train, predictions_train_final)
r2_test = r2_score(target_test, predictions_test)
rmse_train = float(np.sqrt(mean_squared_error(target_train, predictions_train_final)))
rmse_test = float(np.sqrt(mean_squared_error(target_test, predictions_test)))

residuals_test = (target_test - predictions_test).ravel()

# 1) Training cost vs iteration (multivariate GD)
fig_cost_multi, ax_cost_multi = plt.subplots(figsize=(7, 3.5))
ax_cost_multi.plot(training_cost_history, color="steelblue")
ax_cost_multi.set_xlabel("iteration")
ax_cost_multi.set_ylabel(r"$J(\theta)$ (train)")
ax_cost_multi.set_title("Multivariate model: training cost vs iteration")
plt.tight_layout()
plt.show()

# 2) R^2 and RMSE (train vs test) — separate axes (different scales)
fig_metrics, axes_metrics = plt.subplots(1, 2, figsize=(8, 3.5))
axes_metrics[0].bar(["train", "test"], [r2_train, r2_test], color=["steelblue", "coral"])
axes_metrics[0].set_ylabel(r"$R^2$")
axes_metrics[0].set_title(r"Multivariate $R^2$")
axes_metrics[1].bar(["train", "test"], [rmse_train, rmse_test], color=["steelblue", "coral"])
axes_metrics[1].set_ylabel("RMSE")
axes_metrics[1].set_title("Multivariate RMSE")
plt.tight_layout()
plt.show()

print(f"R^2 train: {r2_train:.6f}  |  R^2 test: {r2_test:.6f}")
print(f"RMSE train: {rmse_train:.6f}  |  RMSE test: {rmse_test:.6f}")

# 3) Residual diagnostics (held-out test set)
fig_resid, axes_resid = plt.subplots(1, 2, figsize=(10, 4))
axes_resid[0].scatter(predictions_test.ravel(), residuals_test, s=12, alpha=0.4)
axes_resid[0].axhline(0.0, color="gray", lw=1)
axes_resid[0].set_xlabel("predicted close_next (test)")
axes_resid[0].set_ylabel("residual (actual - predicted)")
axes_resid[0].set_title("Residuals vs fitted (test)")
axes_resid[1].hist(residuals_test, bins=30, edgecolor="black", alpha=0.75)
axes_resid[1].set_xlabel("residual")
axes_resid[1].set_ylabel("count")
axes_resid[1].set_title("Residual histogram (test)")
plt.tight_layout()
plt.show()

# 4) Teaching: 2D slice of cost — vary intercept & first weight; hold other weights at GD solution
def train_cost_slice(parameter_vector):
    err = design_train_multi @ parameter_vector.reshape(-1, 1) - target_train
    return float((err.T @ err) / (2 * n_train_samples))


intercept_hat = float(parameters_multi[0, 0])
first_weight_hat = float(parameters_multi[1, 0])
slice_span_intercept = max(5.0, abs(intercept_hat) * 0.35)
slice_span_w1 = max(0.5, abs(first_weight_hat) * 0.35)
intercept_line = np.linspace(
    intercept_hat - slice_span_intercept, intercept_hat + slice_span_intercept, 80
)
first_weight_line = np.linspace(
    first_weight_hat - slice_span_w1, first_weight_hat + slice_span_w1, 80
)
intercept_grid, first_weight_grid = np.meshgrid(intercept_line, first_weight_line)
cost_slice = np.empty_like(intercept_grid)
base = parameters_multi.ravel().copy()
for i in range(intercept_grid.shape[0]):
    for j in range(intercept_grid.shape[1]):
        trial = base.copy()
        trial[0] = intercept_grid[i, j]
        trial[1] = first_weight_grid[i, j]
        cost_slice[i, j] = train_cost_slice(trial)

fig_slice, ax_slice = plt.subplots(figsize=(6.5, 5))
cs_slice = ax_slice.contour(intercept_grid, first_weight_grid, cost_slice, levels=20)
fig_slice.colorbar(cs_slice, ax=ax_slice, label=r"$J(\theta)$ slice (train)")
ax_slice.plot(intercept_hat, first_weight_hat, "r*", markersize=14, label="GD solution")
ax_slice.set_xlabel(r"$\theta_0$ (intercept)")
ax_slice.set_ylabel(r"$\theta_1$ (weight on standardized open)")
ax_slice.set_title("2D slice of cost (other weights fixed at GD)")
ax_slice.legend()
plt.tight_layout()
plt.show()
```
